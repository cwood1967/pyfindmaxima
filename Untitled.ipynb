```python
import time

import numpy as np
from matplotlib import pyplot as plt
from simr_io.nd2read import nikonImage
from findmaxima import findmaxima
```

```python
imagefile = "/Users/cjw/Desktop/Mitosis/nd2/Top1-SC2--GapR-Nhp2-006.nd2"
nd2 = nikonImage(imagefile)
image = nd2()
image.shape
```

```python
dxy = dict()
times = list()
t0 = time.time()
for i in range(image.shape[0]):
    t1 = time.time()
    dxy[i] = findmaxima.findmaxima(image[i, 1,:,:], 800)
    times.append(time.time() - t1)
    
print("all", time.time() - t0)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(image[8, 1, : :], cmap="gray")
plt.scatter(dxy[8][1], dxy[8][0], s=42, marker="o", color='yellow', alpha=.5)
plt.axis('off')
```
